## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
df2 = pd.read_csv('Resources/client_dataset.csv')
```

```python
df2.head()
```

```python
# Create a column that calculates the 
# subtotal for each line using the unit_price
# and the qty

df2['line_subtotal'] = (df2['unit_price'] * df2['qty'])
df2.head()

```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound
# for orders over 50 pounds and $10 per
# pound for items 50 pounds or under.
def pricer(x):
     if x>50:
        return 7.00
     else:
        return 10.00  
     
x = (df2['unit_weight'] * df2['qty'])
df2['shipping_price'] = x.apply(pricer) * (df2['unit_weight'] * df2['qty'])
df2.head()
```

```python
# Create a column for the total price
# using the subtotal and the shipping price
# along with a sales tax of 9.25%
df2['line_price'] = (df2['line_subtotal'] + df2['shipping_price']) * .0925 + (df2['line_subtotal'] + df2['shipping_price'])
df2.head().round(2)
```

```python
# Create a column for the cost
# of each line using unit cost, qty, and
# shipping price (assume the shipping cost
# is exactly what is charged to the client).
df2['line_cost'] = (df2['unit_cost'] * df2['qty']) +  df2['shipping_price']
df2.head()

```

```python
# Create a column for the profit of
# each line using line cost and line price
df2['line_profit'] = (df2['line_price'] - df2['line_cost'])
df2.head().round(2)
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
top5_df = pd.Series(top5)
#print(f'{top5_df}')
orders_df2 = df2.loc[(df2["client_id"].isin(top5_df))] 

```

```python
# How much did each of the top 5 clients by quantity
# spend? Check your work from Part 1 for client ids.
def order_sum(x):
    return x.sum()


orders_df3 = orders_df2.groupby(["client_id"]).\
            apply(lambda x: pd.Series({"Total_Sum_of_Orders": order_sum(x["line_price"])}))
orders_df3["Total_Sum_of_Orders"] = orders_df3["Total_Sum_of_Orders"].map('${:,.2f}'.format)
orders_df3
```

```python
# Create a summary DataFrame showing the totals for the
# for the top 5 clients with the following information:
# total units purchased, total shipping price,
# total revenue, and total profit. Sort by total profit.
def order_sum(x):
    return x.sum()


orders_df4 = orders_df2.groupby(["client_id"]).\
            apply(lambda x: pd.Series({"qty": order_sum(x["qty"]),                            
                                       "shipping_price": order_sum(x["shipping_price"]),
                                       "line_price": order_sum(x["line_price"]),
                                       "line_cost": order_sum(x["line_cost"]),
                                       "line_profit": order_sum(x["line_profit"])}))
orders_df4["qty"] = orders_df4["qty"].map('{:,.0f}'.format)
orders_df4["shipping_price"] = orders_df4["shipping_price"].map('${:,.2f}'.format)
orders_df4["line_price"] = orders_df4["line_price"].map('${:,.2f}'.format)
orders_df4["line_cost"] = orders_df4["line_cost"].map('${:,.2f}'.format)
orders_df4["line_profit"] = orders_df4["line_profit"].map('${:,.2f}'.format)
orders_df5 = orders_df4.sort_values(by=["line_profit",],ascending=False)
orders_df5

```

```python
# Sort the updated data by "Total Profit" form highest to lowest


def order_sum(x):
    return x.sum()


orders_df8 = orders_df2.groupby(["client_id"]).\
            apply(lambda x: pd.Series({"Units": order_sum(x["qty"]),                            
                                       "shipping_price": order_sum(x["shipping_price"]/1000000),
                                       "line_price": order_sum(x["line_price"]/1000000),
                                       "line_cost": order_sum(x["line_cost"]/1000000),
                                       "line_profit": order_sum(x["line_profit"]/1000000)}))
orders_df9 = orders_df8.rename(columns={"shipping_price": "Shipping", "line_price": "Total Revenue", \
                                        "line_cost": "Total Cost", \
                                        "line_profit": "Total Profit"}).map('${:,.2f}M'.format)
orders_df10 = orders_df9.sort_values(by=["Total Profit",],ascending=False)
orders_df10
```
